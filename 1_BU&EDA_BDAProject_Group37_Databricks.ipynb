```python
# Auxiliary function to display multiple dataframes side by side
# Source: https://python.plainenglish.io/displaying-multiple-dataframes-side-by-side-in-jupyter-lab-notebook-9a4649a4940
from IPython.display import display_html
from itertools import chain,cycle
def display_side_by_side(*args, super_title, titles=cycle([''])):
    """ Display multiple DataFrames side by side in a Jupyter notebook.
    Args:
        args: Variable number of DataFrame objects to be displayed side by side.
        super_title: The main title to be displayed at the top of the combined view.
        titles: An iterable containing titles for each DataFrame to be displayed. Defaults to an infinite cycle of empty strings.

    Returns:
        None. The function generates and displays HTML content side by side for given DataFrames.
    """
    html_str = ''
    html_str += f'<h1 style="text-align: left; margin-bottom: -15px;">{super_title}</h1><br>'
    html_str += '<div style="display: flex;">'
    for df, title in zip(args, chain(titles, cycle(['</br>']))):
        html_str += f'<div style="margin-right: 20px;"><h3 style="text-align: center;color:#555555;">{title}</h3>'
        html_str += df.to_html().replace('table', 'table style="display:inline; margin-right: 20px;"')
        html_str += '</div>'
    html_str += '</div>'
    display_html(html_str, raw=True)
```

## üîé **Missing Values and Distinct Values**

Checking the number of **`null`** values or **`NAs`** in the columns.

```python
# Check missing values and distinct values for reviews_df

# Count the number of distinct values in each column
count_obs = reviews_df.count()

# List to store columns with missing values
missing_values_reviews_df = []

# Define and print the header for the output table
header = "{:<38} | {:<33} | {:<26} | {:<18}".format("\033[1mColumn Name\033[0m", "\033[1mColumn Type\033[0m", "\033[1mEmpty/Null Count\033[0m", "\033[1mPercentage\033[0m")
print(header)
print("-" * len(header)) # Dynamic separator line length

# Iterate through each column structure (name and type) in the DataFrame schema
for col_struct in reviews_df.schema:
    col_name = col_struct.name
    col_type = col_struct.dataType
    col_type_str = str(col_type) # Get a string representation for printing

    # Determine the condition for 'empty/missing' based on the data type
    if isinstance(col_type, StringType):
        # For String columns: check for NULL OR empty string OR whitespace-only string
        # F.trim removes leading/trailing whitespace. If the result is '', it was empty or just whitespace.
        condition = F.col(col_name).isNull() | (F.trim(F.col(col_name)) == '')
        type_desc = "String"

    elif isinstance(col_type, ArrayType):
        # For Array columns (like array<string> or array<struct>): check for NULL OR an empty array (size 0)
        condition = F.col(col_name).isNull() | (F.size(F.col(col_name)) == 0)
        # Make type description more informative for arrays
        type_desc = f"Array({str(col_type.elementType)})"[:25] # Truncate if too long

    else:
        # Default for other types (Numeric, Boolean, BigInt, Struct, etc.): only check for NULL values
        condition = F.col(col_name).isNull()
        type_desc = col_type_str[:25] # Use string representation, truncate if long

    # Count the number of rows where the condition is true for the current column
    empty_count = reviews_df.where(condition).count()

    # Calculate the percentage
    percentage = (empty_count * 100 / count_obs) if count_obs > 0 else 0
    percentage_str = f"{percentage:>9.2f}%"

    # Append the column name and type to the list of missing values
    missing_values_reviews_df.append((col_name, type_desc, empty_count, percentage))

    # Print the results for the current column using f-string formatting
    # \033[1m...\033[0m adds bold formatting for the column name in supporting terminals
    print(f"\033[1m{col_name:<30}\033[0m | {type_desc:<25} | {str(empty_count):<18} | {percentage_str}")
```

## üîß **Data Pre-Proccessing**

```python
# Data Cleaning: Replace Missing/Empty Values with Null

# Define a function to replace identified missing/empty values with null
def replace_missing_with_null(df):
    """
    Replaces missing or empty values in a DataFrame with null based on column type.

    For StringType columns, it replaces:
    - null values
    - empty strings ('')
    - strings containing only whitespace

    For ArrayType columns, it replaces:
    - null values
    - empty arrays (size 0)

    For all other column types, it replaces:
    - null values

    Args:
        df (DataFrame): The input PySpark DataFrame.

    Returns:
        DataFrame: A new DataFrame with missing/empty values replaced by null.
    """
    # Iterate through each column in the DataFrame's schema
    for col_struct in df.schema:
        col_name = col_struct.name
        col_type = col_struct.dataType

        # Determine the condition for 'missing/empty' based on the data type
        if isinstance(col_type, StringType):
            # Condition for String: null, empty, or whitespace only
            condition = F.col(col_name).isNull() | (F.trim(F.col(col_name)) == '')
            print(f"- Applying String null replacement logic to column: {col_name}")

        elif isinstance(col_type, ArrayType):
            # Condition for Array: null or size is 0
            condition = F.col(col_name).isNull() | (F.size(F.col(col_name)) == 0)
            print(f"- Applying Array null replacement logic to column: {col_name}")

        else:
            # Default condition for other types: just null
            condition = F.col(col_name).isNull()
            # Optional: Print only if debugging is needed for other types
            # print(f"- Applying default null replacement logic to column: {col_name} (Type: {col_type})")


        # Apply the replacement using when/otherwise
        # If the condition is met, replace with null (F.lit(None)), otherwise keep the original value
        df = df.withColumn(col_name, F.when(condition, F.lit(None)).otherwise(F.col(col_name)))
    return df

# Apply the function to both DataFrames
reviews_df_cleaned = replace_missing_with_null(reviews_df)
products_df_cleaned = replace_missing_with_null(products_df)
```

```python
# Check if the missing values have been replaced with null (Reviews)
print("\n\033[1mMissing Values in Reviews DataFrame after cleaning:\033[0m")
missing_values_reviews_df_cleaned = []
count_obs = reviews_df_cleaned.count()
header = "{:<38} | {:<33} | {:<26} | {:<18}".format("\033[1mColumn Name\033[0m", "\033[1mColumn Type\033[0m", "\033[1mEmpty/Null Count\033[0m", "\033[1mPercentage\033[0m")
print(header)
print("-" * len(header))
for col_struct in reviews_df_cleaned.schema:
    col_name = col_struct.name
    col_type = col_struct.dataType
    col_type_str = str(col_type)  # Get a string representation for printing
    type_desc = col_type_str[:25] # Use string representation, truncate if long

    # Count the number of rows where the condition is true for the current column
    empty_count = reviews_df_cleaned.where(F.col(col_name).isNull()).count()

    # Calculate the percentage
    percentage = (empty_count * 100 / count_obs) if count_obs > 0 else 0
    percentage_str = f"{percentage:>9.2f}%"

    # Append the column name and type to the list of missing values
    missing_values_reviews_df_cleaned.append((col_name, type_desc, empty_count, percentage))

    # Print the results for the current column using f-string formatting
    print(f"\033[1m{col_name:<30}\033[0m | {type_desc:<25} | {str(empty_count):<18} | {percentage_str}")
```

```python
# Check if the missing values have been replaced with null (Products)
print("\n\033[1mMissing Values in Products DataFrame after cleaning:\033[0m")
missing_values_products_df_cleaned = []
count_obs = products_df_cleaned.count()
print(header)
print("-" * len(header))
for col_struct in products_df_cleaned.schema:
    col_name = col_struct.name
    col_type = col_struct.dataType
    col_type_str = str(col_type) # Get a string representation for printing
    type_desc = col_type_str[:25] # Use string representation, truncate if long

    # Count the number of rows where the condition is true for the current column
    empty_count = products_df_cleaned.where(F.col(col_name).isNull()).count()

    # Calculate the percentage
    percentage = (empty_count * 100 / count_obs) if count_obs > 0 else 0
    percentage_str = f"{percentage:>9.2f}%"

    # Append the column name and type to the list of missing values
    missing_values_products_df_cleaned.append((col_name, type_desc, empty_count, percentage))

    # Print the results for the current column using f-string formatting
    print(f"\033[1m{col_name:<30}\033[0m | {type_desc:<25} | {str(empty_count):<18} | {percentage_str}")
```

#### **Verification of Missing Value Cleaning:**
This section applies the `replace_missing_with_null` function to standardize various forms of missing or empty data across both DataFrames. Specifically, it converts:
*   NULLs, empty strings (`''`), and whitespace-only strings to `NULL` for `StringType` columns.
*   NULLs and empty arrays/structs (size 0) to `NULL` for `ArrayType` or `StructType` columns.
*   Only existing NULLs are considered for other data types (Numeric, Boolean, etc.).

The subsequent missing value check confirms the cleaning process by re-calculating the null counts, **now correctly identifying all standardized missing entries using `.isNull()`**. This provides a unified view of missing data after accounting for empty strings and arrays which were previously treated separately.
```python
# Users that reviewed different products (user_id) - 'user_id' (number of unique users that reviewed the product)
# 2 Tables (1st 1 - max(reviews that same user reviewed different products) | 2nd 0, >0, >1, >2, ..., max(reviews that same user reviewed different products)
# Display the tables side by side

# 1. Calculate the number of distinct products (parent_asin) reviewed by each user
user_distinct_product_counts = reviews_df.groupBy("user_id") \
    .agg(F.countDistinct("parent_asin").alias("num_distinct_products"))

# 2. Calculate Total Number of Unique Users who reviewed at least one product
total_users = user_distinct_product_counts.count()
print(f"\n\033[1mTotal Unique Users who reviewed products:\033[0m {total_users}")

# 3. Prepare Table 1: Frequency of users reviewing *exactly* X distinct products
table1_df = user_distinct_product_counts \
    .groupBy("num_distinct_products") \
    .agg(F.count("*").alias("n Users")) \
    .withColumn("% Users", F.round((F.col("n Users") / total_users * 100), 2)) \
    .orderBy("num_distinct_products") \
    .select(
        F.col("num_distinct_products").alias("n Distinct Products Reviewed"),
        F.col("n Users"),
        F.col("% Users")
    )

# 4. Prepare Table 2: Frequency of users reviewing *more than* X distinct products
#    Calculate cumulative count of users reviewing >= X products (descending order)
w_desc = Window.orderBy(F.col("n Distinct Products Reviewed").desc()) \
                .rowsBetween(Window.unboundedPreceding, Window.currentRow)
table2_intermediate = table1_df \
    .withColumn("n Users_ge_X", F.sum("n Users").over(w_desc))

# Filter to keep only the thresholds up to "> 10" (which corresponds to N >= 11)
# We want to keep rows where 'n Distinct Products Reviewed' (N) is 1, 2, ..., 11.
# The row N=11 contains the count for >= 11, which is what we want for "> 10".
table2_filtered = table2_intermediate.filter(F.col("n Distinct Products Reviewed") <= 11)

# The count for "> (X-1)" is the same as ">= X"
# We need thresholds "> 0", "> 1", "> 2", ...
# So, for row N (n Distinct Products Reviewed=N), n Users_ge_X gives count(>=N), which is count(> N-1)
table2_df = table2_filtered \
    .select(
        # The threshold "> (N-1)" is derived from num_distinct_products_reviewed (N)
        (F.col("n Distinct Products Reviewed") - 1).alias("threshold_value"),   # Creates 0, 1, ..., 10
        F.col("n Users_ge_X").alias("n Users")                                  # This is the correct count for >=N (or >N-1)
    ) \
    .withColumn("reviews_threshold", F.concat(F.lit("> "), F.col("threshold_value").cast("string"))) \
    .withColumn("% Users", F.round((F.col("n Users") / total_users * 100), 2)) \
    .select("reviews_threshold", "n Users", "% Users") \
    .orderBy(F.col("threshold_value"))                                          # Order by 0, 1, 2...

# Display the results
table1_pd = table1_df.toPandas()
table1_pd.set_index('n Distinct Products Reviewed', inplace=True)
table1_pd.index.name = 'n Distinct Products Reviewed'
table2_pd = table2_df.toPandas()
table2_pd.set_index('reviews_threshold', inplace=True)
table2_pd.index.name = 'n Reviews'
display_side_by_side(table1_pd, table2_pd, titles=['Table 1: Exact Frequency', 'Table 2: Cumulative Frequency'], super_title='Users reviewing different products')
```

```python
# Absolute and Relative frequency of reviews text length (0 , 1 - 10 | 11 - 50 | 51 - 100 | 101 - 250 | 251 - 500 | 501+ words)
# 1. Calculate word count
reviews_with_word_count = reviews_df.withColumn(
    "review_text_word_count",
    F.when(F.col("text").isNull() | (F.trim(F.col("text")) == ""), 0).otherwise(
        F.size(F.split(F.trim(F.col("text")), "\s+"))
    )
)

# 2. Categorize based on word count
reviews_categorized = reviews_with_word_count.withColumn(
    "review_text_length_category",
    F.when(F.col("review_text_word_count") == 0, "0 words")
     .when(F.col("review_text_word_count").between(1, 10), "1 - 10 words")
     .when(F.col("review_text_word_count").between(11, 50), "11 - 50 words")
     .when(F.col("review_text_word_count").between(51, 100), "51 - 100 words")
     .when(F.col("review_text_word_count").between(101, 250), "101 - 250 words")
     .when(F.col("review_text_word_count").between(251, 500), "251 - 500 words")
     .when(F.col("review_text_word_count") >= 501, "501+ words")
     .otherwise("Unknown")
)

# 3. Calculate total count for relative frequency calculation
total_count = reviews_categorized.count()

# 4. Calculate absolute and relative frequencies
frequency_reviews_df = (reviews_categorized
                         .groupBy("review_text_length_category")
                         .agg(F.count("*").alias("n"))
                         .withColumn("%", F.round((F.col("n") / total_count), 4))
                        )

# 5. Order the results logically
category_order = ["0 words", "1 - 10 words", "11 - 50 words", "51 - 100 words", "101 - 250 words", "251 - 500 words", "501+ words"]
frequency_reviews_df = frequency_reviews_df.withColumn(
    "category_sort_order",
    F.when(F.col("review_text_length_category") == "0 words", 0)
     .when(F.col("review_text_length_category") == "1 - 10 words", 1)
     .when(F.col("review_text_length_category") == "11 - 50 words", 2)
     .when(F.col("review_text_length_category") == "51 - 100 words", 3)
     .when(F.col("review_text_length_category") == "101 - 250 words", 4)
     .when(F.col("review_text_length_category") == "251 - 500 words", 5)
     .when(F.col("review_text_length_category") == "501+ words", 6)
     .otherwise(7) # For any unexpected categories
).orderBy("category_sort_order").drop("category_sort_order")

# Show the results
frequency_reviews_pandas_df = frequency_reviews_df.toPandas()
frequency_reviews_pandas_df.set_index('review_text_length_category', inplace=True)
frequency_reviews_pandas_df.index.name = "Review Text Length Category"
frequency_reviews_pandas_df['%'] = frequency_reviews_pandas_df['%'] * 100
frequency_reviews_pandas_df['Cumulative %'] = frequency_reviews_pandas_df['%'].cumsum()
frequency_reviews_pandas_df
```

###### **8 | `timestamp`**: Time of the review (unix time)

```python
# Absolute and Relative frequency of reviews per year, month, day of the week and hour (4 dataframes with display_side_by_side - 'timestamp')

# 1. Calculate Total Number of Reviews
total_reviews = reviews_df.count()
print(f"\n\033[1mTotal Number of Reviews:\033[0m {total_reviews}")

# 2. Calculate Frequency per Year
freq_year_df = reviews_df.groupBy("year") \
    .agg(F.count("*").alias("n Reviews")) \
    .withColumn("%", F.round((F.col("n Reviews") / total_reviews * 100), 2)) \
    .orderBy("year")
freq_year_pd = freq_year_df.toPandas().set_index('year') # Set year as index for display

# 3. Calculate Frequency per Month
freq_month_df = reviews_df.groupBy("month") \
    .agg(F.count("*").alias("n Reviews")) \
    .withColumn("%", F.round((F.col("n Reviews") / total_reviews * 100), 2)) \
    .orderBy("month")
freq_month_pd = freq_month_df.toPandas().set_index('month') # Set month as index

# Map month number to name in Pandas for better readability
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
              7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
freq_month_pd.index = freq_month_pd.index.map(month_map)
freq_month_pd.index.name = 'month_name'

# 4. Calculate Frequency per Day of the Week
# Spark's dayofweek: 1=Sun, 2=Mon, ..., 7=Sat
freq_dow_df = reviews_df.groupBy("day_of_week") \
    .agg(F.count("*").alias("n Reviews")) \
    .withColumn("%", F.round((F.col("n Reviews") / total_reviews * 100), 2)) \
    .orderBy("day_of_week") # Order by the numeric representation first
freq_dow_pd = freq_dow_df.toPandas()

# Map numeric day_of_week to names and set desired order (e.g., Mon-Sun) in Pandas
dow_map = {1: 'Sun', 2: 'Mon', 3: 'Tue', 4: 'Wed', 5: 'Thu', 6: 'Fri', 7: 'Sat'}
desired_dow_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
freq_dow_pd['day_name'] = freq_dow_pd['day_of_week'].map(dow_map)
freq_dow_pd = freq_dow_pd.set_index('day_name').loc[desired_dow_order].drop(columns=['day_of_week']) # Set index, reorder, drop numeric col

# 5. Calculate Frequency per Hour
freq_hour_df = reviews_df.groupBy("hour") \
    .agg(F.count("*").alias("n Reviews")) \
    .withColumn("%", F.round((F.col("n Reviews") / total_reviews * 100), 2)) \
    .orderBy("hour")
freq_hour_pd = freq_hour_df.toPandas().set_index('hour') # Set hour as index

# 6. Display the results side-by-side using the provided function
display_side_by_side(freq_year_pd, freq_month_pd, freq_dow_pd, freq_hour_pd,
                     super_title="Review Frequency Analysis by Time",
                     titles=['By Year', 'By Month', 'By Day of Week', 'By Hour of Day'])
```
